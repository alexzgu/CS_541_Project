```python
import pandas as pd

pred_path = 'segment_break.csv'
ground_truths_dir = '../../data/clean/segment_breaks'

pred_df = pd.read_csv(pred_path)
```

```python
# iterate through the ground_truths_dir and check the number of rows in each csv
import os
import pandas as pd

ground_truths = os.listdir(ground_truths_dir)
ground_truths = [f for f in ground_truths if f.endswith('.csv')]

for f in ground_truths:
    df = pd.read_csv(os.path.join(ground_truths_dir, f))
    # last end value
    last_end = df['end'].iloc[-1]
    print(f, last_end)
```
